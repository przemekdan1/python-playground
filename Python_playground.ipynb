## **Program zliczający wszystkie słowa i wystąpienia wybranego**

```python
def count_words(string, word_to_search):
    words = string.split()  # splits string into words
    counter = 0
    for word in words:
        if word == word_to_search:
            counter += 1
    return len(words), counter
```

##**Klasy abstrakcyjne**

- na górze hierarchii dziedziczenia znajdują się **klasy "bazowe"** - ABC (Abstract Base Classes)
- nie chcemy aby były do nich tworzone instancje
  - mamy ogólną klasę Employee i konkretne klasy, np. Programmer, Manager, Janitor
  - nigdy nie chcemy tworzyć ogólnej klasy, bo nikt nie jest "samym" Employee
  - gdy trzeba nowych typów, to tworzymy nowe klasy dziedziczące
  - klasa Employee **(abstrakcyjna)** zawiera wspólne atrybuty i metody
- istnieją dla programisty, dla lepszej oraganizaji kodu
- łączą wspólne elementy klas, na wyzszym poziomie
- podkreślenie, ze z tej klasy się dziedziczy
- wyznaczenie ABC uniemozliwia stworzenie jej instancji


```python
from abc import ABC

class Employee(ABC):
    def __init__(self, name, ...):
        self.name = name

class Manager(Employee):
    pass

e = Employee("Jan", "Kowalski")  # error!
m = Manager("Jan", "Kowalski")   # ok
```

- klasy abstrakcyjne mogą mieć 2 rodzaje metod:
  - zwykłe metody (concrete methods)
  - metody abstrakcyjne (***@abstractmethod***)

**Metody abstrakcyjne**  
- słuzące do nadpisywania przez klasy dziedziczące, **MUSZĄ JĄ NADPISAĆ**
- stosujemy je gdy w 100% chcemy aby klasy dziedziczące miały jakąś metodę, a nie da się zrobić domyślnej implementacji


```python
#nie da się zdefiniować metody poniewaz dla kazdego pracownika sposób jej wyliczania będzie inny, dlatego stosujemy @abstractmethod

from abc import abstractmethod

class Employee:
    # implementation from previous slides

    @abstractmethod
    def calculate_net_pay(self):
        raise NotImplementedError()
```

- tworzą **interfejs** klas dziedziczących
- czasem nie chcemy w ogóle żadnej domyslnej implementacji - **interfejs**
- ABC zawiera tylko metody abstrakcyjne, narzuca implementację metod: "*klasy liczące mają mieć metodę tą,tą i tą*"
- klasy dziedziczące muszą implementować klasy abstrakcyjne !!!

```python
def decorator_name(fun):              #po tej nazwie odwołujemy się do dekoratora, np. @decorator_name, wywoływanie

    @functools.wraps(fun)             #zapewnia transparentność dekoratora, wskazywanie Pythonowi ze to dalej ta sama funkcja
    def wrapper(*args, **kwargs):     #"udekorowana" funkcja
        # before fun
        val = fun(*args, **kwargs)    #wywołanie dekorowanej funkcji, funkcja z zewnątrz
        # after fun
        return val
    return wrapper                    #zwrócenie obiektu udekorowanej funkcji
```

###**Przykładowe zastosowanie dekoratorów**

**Mierzenie czasu wykonania funkcji**

```python
def measure_time(fun):
    """ Prints the runtime """

    @functools.wraps(fun)
    def wrapper(*args, **kwargs):
        start = time.time()               #początek funkcji
        value = fun(*args, **kwargs)
        end = time.time()                 #koniec funkcji
        print("Time:", end - start)
        return value

    return wrapper

@measure_time
def manhattan_dist(x, y):                 #metryka manhattanu
    elem = lambda a, b: abs(a - b)
    dist = sum(map(elem, x, y))
    return dist
```

```python
import functools

class CountCalls:
    def __init__(self, fun):
        functools.update_wrapper(self, fun)                           #funkcja jest dalej funkcją
        self.fun = fun                                                #obiekt funkcji
        self.num_calls = 0                                            #liczba wywołań, stateful decorator

    def __call__(self, *args, **kwargs):
        self.num_calls += 1
        print(f"Call {self.num_calls} of {self.fun.__name__!r}")
        return self.fun(*args, **kwargs)
```

###**Generator functions**

- gdy do funkcji dodamy w którymkolwiek miejscu yield, to staje się generator function

```python
def fibonacci():
    x, y = 0, 1
    while True:
        x, y = y, x + y
        yield x

if __name__ == '__main__':
    fib = fibonacci()
    print(next(fib))  # 1
    print(next(fib))  # 1
    print(next(fib))  # 2

    fib2 = fibonacci()  # 1, 1, 2, ...
```

##**Pipeline pattert**

- plan w jakiej kolejności mają się wykonać obiekty funkcji

####**Implementacja**

- Pipeline - klasa zbierająca operacje do wykonania jako callable
- PipelineConfig - konfiguracja (często dataclass) pipeline'u, dodatkowe argumenty funkcji
- kroki przetwarzania - implementują określony interfejs (funkcje lub klasy z __call__())
- wyjście poprzedniej funkcji (typowo ramka z danymi / lista danych) to wejście następnej


```python
#inplementacja
from functools import partial, reduce
from typing import Callable


class Pipeline:
    def __init__(self):
        self._steps = []                                                          #lista kroków do wykonania

    def add_step(self, fun: Callable, **kwargs) -> None:
        if len(kwargs) > 0:                                                       #gdy przekazany został jakiś argument
            fun = partial(fun, **kwargs)
        self._steps.append(fun)

    def execute(self, input_data):
        pipeline_fun = lambda prev_output, next_fun: next_fun(prev_output)

        return reduce(pipeline_fun, self._steps, input_data)
```

##**--- REST API ---**

###**HTTP**

####**Kody odpowiedzi HTTP**

**Informacyjne - 1xx**

**Powodzenia - 2xx**
- 200 - **OK**
- 201 - **Created** - dokument został pomyślnie stowrzony przez serwer
- 204 - **No Content** zapytanie zrealizowane, ale bez zadnej zwracanej zadnej treści

**Przekierowania - 3xx**
- 301 - **Moved Permanently** - zapytanie przekierowane na inny adres url
- 304 - **Not Modified** - klient moze uzwyać zcacheowanej wersji odpowiedzi

**Błędu klienta - 4xx**
- 400 - **Bad Request** - serwer nie jest w stanie odczytać zapytania klienta
- 401 - **Unauthorized** - zasób dostępny tylko dla zalogowanych uzytkowników
- 403 - **Frobidden** - serwer zrozumiał zapytanie, ale ze względów bezpieczeństwa nie moze zwrócić odpowiedzi
- 404 - **Not Found** - zasób nie istnieje na serwerze

**Błędu serwera - 5xx**
- 500 - **Internal Server Error** - bug na serwerze

####**Podstawowe metody HTTP**

- **GET** - pobieranie zasobu
- **POST** - przesłanie zasobu
- **PUT/PATCH** - aktualizacja zasobu
- **DELETE** - usunięcie zasobu

##**--- Wprowadzenie do sieci neuronowych - Tensorflow 2.0 + Keras ---**

###**Słowniczek pojęć**

- **perceptron** - struktura nieposiadająca warstw ukrytych

###**Podział rodzajów uczenia maszynowego**

- **supervised learnig**
  - klasyfikacja (przydzielenie do odpowiedniej grupy)
  - regresja (za pomocą podobnych parametrów dokonujemy predykcji)
- **unsupervised learnig**
  - klasteryzacja (przydzialanie do klastrów)
  - reguły asocjacyjne (np. co jest kupowane z jakimś produktem, pieluchy+piwo)
- **semi-supervised learnig** (przewidywanie niekompletych zestawów danych)
- **reinforcement learning** (kazemy lub wynagadzamy model za decyzje, metoda prób i błędów, model stara się unikac kar)

###**Playground tensorflow**

####**Opcje**

- **learning rate** - szybkość dopasowania modelu,
- **activation** - funkcja aktywacji
- **regularization** - regularyzacja(L1/L2)
- **problem type** - określenie problemu (klasyfikacja/regresja)
- podział danych na treningowe i walidacyjne
- **noise** - okreslanie poziomu szumu
- **batch size** - rozmiar wsadu po którym zostaną aktualizowane dane w sieci
- wybór typu danych:
  - liniowo rozdzielne
  - dane typu XOR
  - dane sferyczne
  - dane spiralne
